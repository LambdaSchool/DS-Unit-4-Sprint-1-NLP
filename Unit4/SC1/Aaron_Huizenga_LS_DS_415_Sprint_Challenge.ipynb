## Part 1: Tokenize Function
<a id="#p1"></a>

Complete the function `tokenize`. Your function should
- accept one document at a time
- return a list of tokens

You are free to use any method you have learned this week.

```
import spacy

nlp = spacy.load('en_core_web_sm')
```

```
# this will define the stop_words

stop_words = nlp.Defaults.stop_words
```

```
# Need to create a function to do a tokenization test

def tokenize(docs):
    for tokens in docs:
        
        text = [tokens.lemma_ for tokens in nlp(docs) if
               (tokens.is_punct == False) and
               (tokens.text.lower() not in stop_words)]
        
    return text
```

## Part 2: Vector Representation
<a id="#p2"></a>
1. Create a vector representation of the reviews
2. Write a fake review and query for the 10 most similiar reviews, print the text of the reviews. Do you notice any patterns?
    - Given the size of the dataset, it will probably be best to use a `NearestNeighbors` model for this. 

```
# Gotta start by creating a TF-IDF matrix

from sklearn.feature_extraction.text import TfidfVectorizer

def tfdif_vector(docs):
    # now to instantiate vectorizer

    tfdf = TfidfVectorizer(stop_words = 'english', 
                           ngram_range = (1, 2),
                           min_df = 3,
                           max_df = 0.5)
    
    # I need to create a vocab and tf-idf score for each doc
    tf_score = tfdf.fit_transform(docs)
    
    # This will pull feature names and use them as column headers
    tf_score = pd.DataFrame(tf_score.todense(),
                            columns = tfdf.get_feature_names())
    
    shape = print(f'tf_score shape = {tf_score.shape}')

    head = tf_score.head()

    return shape, head, tf_score, tfdf
```

```
# This will hopefully vectorize the text column from the yelp dataset
# In essence, this is just a test to make sure that the function works properly.

tfdif_vector(yelp['text'])
```

```
# Now that you can see it is working, I will use the function but assign variables
# to it here.

tf_shape, tf_head, doc_vector, tfdf = tfdif_vector(yelp['text'])
```

## Part 3: Classification
<a id="#p3"></a>
Your goal in this section will be to predict `stars` from the review dataset. 

1. Create a piepline object with a sklearn `CountVectorizer` or `TfidfVector` and any sklearn classifier. Use that pipeline to estimate a model to predict `stars`. Use the Pipeline to predict a star rating for your fake review from Part 2. 
2. Tune the entire pipeline with a GridSearch

```
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```
# I need to instantiate RFC

random = RandomForestClassifier()
```

```
# Next - the pipeline needs to be defined
# I will use the vector already defined in the tfdif vect function

pipe = Pipeline([
       ('vect', tfdf),
       ('clf', random)
])
```

```
# In a traditional move, I will take this info and split into train and test 
# dataset.

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(yelp['text'],
                                                    yelp['stars'],
                                                    test_size=0.2,
                                                    stratify=yelp['stars'],
                                                    random_state=42)
```

```
# Now I need to fit the pipeline to the train set

pipe.fit(x_train, y_train)
```

```
from sklearn.metrics import accuracy_score

# This will result in the accuracy score

y_pred = pipe.predict(x_test)
accuracy_score(y_test, y_pred)
```

```
# I will use the pipeline above to predict the star rating for my sample review.
# Although our pipeline gave us a rating of 4 it was actually a 5.

pipe.predict(review)
```

```
# I will use GridSearchCV to find the most optimal params for my model

param = {
    'vect__max_df': (0.5, 1.0),
    'vect__min_df': (2, 5, 10),
    'vect__max_features': (500, 10000),
    'clf__n_estimators': (5, 20),
    'clf__max_depth': (10, 50, None)
}

grid_search = GridSearchCV(pipe, param, cv = 5, n_jobs=-1, verbose=1)
grid_search.fit(x_train, y_train)
```

```
# This will print out the best scores from the GridSearch

grid_search.best_score_
```

```
# Let's take a look at the best params

grid_search.best_params_
```

```
# Take 2, only this time I'll use the best params from the search

random2 = RandomForestClassifier(max_depth = None,
                                 n_estimators = 20)

tfdf2 = TfidfVectorizer(stop_words = 'english',
                          ngram_range = (1, 2),
                          min_df = 2,
                          max_features = 10000)

pipe = Pipeline([
    ('vect', tfdf2),
    ('clf', random2),
])

pipe.fit(x_train, y_train)

print(f'New Star Prediction {pipe.predict(review)}')

# Accuracy Score

y_pred = pipe.predict(x_test)
accuracy_score(y_test, y_pred)
```

```
# Libraries necessary
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import STOPWORDS
from gensim import corpora
from gensim.models import LdaMulticore
from gensim.corpora import Dictionary

import pandas as pd
from tqdm import tqdm
tqdm.pandas
```

### I am only working with 2 concurrent workers, and since this particular document is 10k rows long, it will take FOREVER to tokenize it. So I will only be using 50 sample rows instead of using it in its entirety.

```
# Working with the sample set

df = yelp.sample(random_state=42, n=50)

print(df.shape)
df.head()
```

```
# I will ad a column to the sample set that has the tokens

df['tokens'] = df['text'].progress_apply(tokenize)

df.head(1)
```

```
# As you can see, even with 50 it still took almost 30min.

# Now I will create a dictionary rep of all words
dct = corpora.Dictionary(df['tokens'])
```

Create a bag of words representation of the entire corpus

```
corpus = [dct.doc2bow(text) for text in df['tokens']]
```

Your LDA model should be ready for estimation: 

```
lda = LdaMulticore(corpus=corpus,
                  id2word = dct,
                  random_state = 42,
                  num_topics=5,
                  passes=10)
```

```
# This will refine the looks of the topics

import re

words = [re.findall(r'"([^"]*)"',t[1]) for t in lda.print_topics()]
topics = [' '.join(t[0:6]) for t in words]

for id, t in enumerate(topics): 
    print(f"------ Topic {id} ------")
    print(t, end="\n\n")
```
