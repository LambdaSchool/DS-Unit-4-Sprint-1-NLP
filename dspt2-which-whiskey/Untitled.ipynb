```python
df_train['description'] = df_train['description'].apply(lambda x: re.sub(r'[^a-zA-Z ]', '', x))
```

```python
df_train.head
```

```python
nlp = spacy.load("en_core_web_lg")
tokenizer = Tokenizer(nlp.vocab)
STOP_WORDS = nlp.Defaults.stop_words.union([
    ' ',
    '  ',
    '   ',
    'like', 
    'with', 
    'aged', 
    'whisky',
    'whiskey',
])
```

```python
# create tokens from reviews text
tokens = []

for doc in tokenizer.pipe(df_train['description'], batch_size=500):
    
    doc_tokens = []
    
    for token in doc: 
        if token.text.lower() not in STOP_WORDS:
            doc_tokens.append(token.lemma_.lower())
            
    tokens.append(doc_tokens)
    
df_train['tokens'] = tokens
```

```python
tokens
```

```python
# The object `Counter` takes an iterable, but you can instaniate an empty one and update it. 
word_counts = Counter()

# Update it based on a split of each of our documents
df_train['tokens'].apply(lambda x: word_counts.update(x))

# Print out the 25 most common words
word_counts.most_common(25)
```
